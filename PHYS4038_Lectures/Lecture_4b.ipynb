```python
import pylab # handy for interactive use
import matplotlib.pyplot as plt

plt.ion() # turn on interactive mode!
```

```python
import numpy as np
from numpy import sin, cos, pi
x = np.arange(0, 2*pi, pi/100)
y = sin(x) * cos(2*x)
plt.plot(x,y)
plt.plot(x, sin(x), '--r')
plt.plot(x, cos(2*x), linestyle='dotted', color='green')

# boolean array, false if y <= 0.75, true if y >= 0.75
thresh = y > 0.75

# x[thresh] picks out bits of x where thresh == True
plt.plot(x[thresh], y[thresh], 'r', linewidth=5)
zeros = np.abs(y) < pi/200
plt.plot(x[zeros], y[zeros], 'ok', markersize=10)
plt.xlabel(r'$x$')
plt.ylabel(r'$\sin(x)\cos(2x)$')
plt.axis([0, 2*pi, -1.1, 1.1])
plt.savefig('wiggles.pdf')
```

## Some useful functions

- figure - create a new figure, or get an existing figure object
- plot - add lines or points
- hist/hist2d - create a 1D/2D histogram
- imshow/contour - plot an array as an image/contour
- axis - set axis limits
- subplots - create new figure with a grid
- subplots_adjust - adjust canvas margins

Some functions update the plot, others don't (for efficiency):
To update the plot display:

- draw() - draw plot and continue
- show() - blocks interpreter until window closed
- close(), close('all') - close figure windows, close all figure windows

# Scatter plot example

```python
#use to embed image in the Jupyter notebook
#%matplotlib inline
#use to make interactive plots in a Jupyter notebook
#%matplotlib notebook
from matplotlib import pyplot as plt
import numpy as np
```

```python
x = np.array([ 1,  3,  9, 14, 15, 19, 20, 21, 30, 35,
              40, 41, 42, 47, 54, 56, 67, 69, 72, 88])

y = np.array([33, 68, 34, 34, 37, 65, 37, 44, 48, 49,
              53, 49, 50, 48, 56, 60, 61, 63, 44, 71])

e = np.array([3.6, 3.9, 2.6, 3.4, 3.8, 14.8, 2.2, 2.1, 2.3, 3.8,
              2.2, 2.8, 3.9, 3.1, 3.4, 2.6, 3.4, 3.7, 2.0, 3.5])
```

```python
plt.plot(x, y)
```

```python
plt.clf() # clear figure
plt.plot(x, y, 'o') # change from line to dots
```

```python
plt.clf()
plt.scatter(x, y) # an alternative, which has more functionality than
                    # plt.plot()
                    # Can change sizes and colours point-by-point
```

```python
plt.clf()
plt.plot(x, y, 'o', linestyle='', color='k')
```

```python
plt.clf()
_ = plt.errorbar(x, y, e, fmt='ok')
```

```python
def model(x):
    return 30 + 0.6*x - 0.002*x**2

xm = np.arange(0, 91, 1)
ym = model(xm)
```

```python
plt.errorbar(x, y, e, fmt='ok')
plt.plot(xm, ym)
plt.xlabel('resistance, $\Omega$ ohms')
plt.ylabel('size of universe, $S$ Gpc')
plt.axis((-10, 100, 20, 90))
```

```python
plt.close()
```

## Plottng - interactive notebook plots

- Some tools for producing itneractive plots in a web browser (via JavaScript), and hence in IPython notebooks

- matplotlib
    - %matplotlib inline - inserts image of plots in notebook
    - %matplotlib notebook - inserts interactive plot in notebook
    
- mpld3
    - use matplotlib commands
    - generate HTML with mpld3 - automatically get pan and zoom
    - optionally add interactivity
    
- bokeh, plotly
    - similar functionality, but different (non-matplotlib) interface
